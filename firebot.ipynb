```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python
usernameStr = 'XXXXXXXXX.com'
passwordStr = 'XXXXXXXXXXXXXXXXX!'
```

```python
browser = webdriver.Chrome()
browser.get(('https://parts.revgroup.com/gensuite/web/login'))
# fill in username and hit the next button
username = browser.find_element_by_id('userName')
username.send_keys(usernameStr)
password = WebDriverWait(browser, 10).until(
    EC.presence_of_element_located((By.ID, 'passwordLogin')))
password.send_keys(passwordStr)

loginButton = browser.find_element_by_id('loginUser')
loginButton.click()

orders = pd.read_csv('hts.csv', dtype='unicode')
orders = orders[['Material']]
#orders = orders[6781:]
description = []
i = 0
for index, row in orders.iterrows():
    # Get the part number to search from the data frane
    element_to_search = row['Material']
    print(element_to_search)
    search = WebDriverWait(browser, 10).until(
        EC.presence_of_element_located((By.NAME, 'keyword')))
    # Search for the search bar and search the part number for the iteration 
    search.send_keys(element_to_search)
    searchButton = browser.find_element_by_class_name('input-group-append')
    searchButton.click()
    time.sleep(5)
    text = browser.find_element_by_tag_name("h5").get_attribute("innerText")
    
    #desc = WebDriverWait(browser, 10).until(
       # EC.presence_of_element_located((By.CLASS_NAME, 'font-weight-semi-bold.ng-binding.ng-scope')))
    description.append(text)
    i = i + 1
    time.sleep(5)
    
orders["Description"] = description
    
```

```python
print(len(description))
x = len(description)
completed_search = orders[:x]
print(completed_search.size)
completed_search["Description"] = description
completed_search.to_csv('completed.csv', index=True, header = True, sep=',')
print(completed_search)
```
